```python
import pandas as pd
data = [{'message':'i like the movie','sentiment':'positive'},
        {'message':'movie is not good','sentiment':'negative'},
        {'message':'wasted my time','sentiment':'negative'},
        {'message':'liked the movie','sentiment':'positive'}
        ]
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
messages = df['message']
label = df['sentiment']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
cv_matrix = cv.fit_transform(messages)
cv_df = pd.DataFrame(cv_matrix.toarray(),columns=cv.get_feature_names_out())
cv_df.head()
```

```python
cv_matrix
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='linear')
```

```python
svc.fit(cv_matrix,label)
```

```python
def predict_text(text):
  cv_text = cv.transform([text])
  return svc.predict(cv_text)
```

```python
result = predict_text('movie is not good')
print(result[0])
```

TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer()
tf_matrix = tf.fit_transform(messages)
tf_df = pd.DataFrame(tf_matrix.toarray(),columns=tf.get_feature_names_out())
tf_df.head()
```

```python
svc.fit(tf_matrix,label)
```

```python
def predict_text_tf(text):
  tf_text = tf.transform([text])
  return svc.predict(tf_text)
```

```python
result = predict_text_tf('movie is not good')
print(result[0])
```
